## In this notebook I import, merge the datasets to create one master dataest with the correct variables for our team to use.

```python
#import python libraries
import pandas as pd
import geopandas as gpd 
```

```python
# read in census data to notebook
census=gpd.read_file ('Data/census.csv')
```

```python
census.head(3)
```

```python
# need to make match id in the census dataset
census['geoid20_x']=census.CT20
```

```python
census.head(3)
```

```python
census.info(verbose=True,show_counts=True)
```

```python
#read in USC grocery data
grocery=gpd.read_file ('Data/USCgrocery.csv', dtype= {'geoid20_x':str})

#drop geometry column
grocery = grocery.drop(['geometry'],axis=1)
```

```python
grocery.head(3)
```

```python
# read in USC data to notebook
usc=gpd.read_file ('Data/usc.csv', dtype= {'geoid20_x':str})

#drop geometry column
usc = usc.drop(['geometry'],axis=1)
```

```python
usc.head(3)
```

```python
usc.info(verbose=True,show_counts=True)
```

```python
#now merge both USC datasets together
#merging usc and census data
merged=grocery.merge(usc, on="geoid20_x")
```

```python
merged.head(3)
```

# now i have both data sets (USC and census) and need to merge on a common match ID, geoid2_x

```python
#merging usc and census data
final=census.merge(merged, on="geoid20_x")
```

```python
final.head(3)
```

```python
final.shape
```

```python
final.info
```

```python
final.info(verbose=True,show_counts=True)
```

```python
new_df=final[["geoid20_x", "CT20", "FIPS_census", "tract","ShapeSTArea", "ShapeSTLength", "geometry_x", "% Hispanic or Latino","% Not a Citizen", "% Unemployed", "Population Density (Per Sq. Mile)",
            "Median Household Income", "Median Family Income", "% Enrolled in School", "% No Health Insurance Coverage", 
              "pop_below_100_pct", "calfresh_per_100_fam_below_200f", "mental_health_pct", "chd_pct", "snap_rate", "spa", "calfresh_per_100_families", "Percent 65 years and over",
            "lowaccess_count", "denom_total_pop", "lowaccess_li_count", "lowaccess_seniors_count" ]]
new_df.head(3)
```

```python
# Export DataFrame to CSV
new_df.to_csv('final.csv', index=False)
```
